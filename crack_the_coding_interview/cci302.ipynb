```python
class StackWithMin:
    def __init__(self, values=None):
        self.stack = []
        self.min = []
        if values!=None:
            self.push(values)
        
    def pop(self):
        if self.is_empty():
            return None
        value = self.stack.pop()
        if self.min[-1] == value:
            self.min.pop()
        return value
    
    def push_single(self, value):
        self.stack.append(value)
        if self.min==[] or self.min[-1]>=value: #in the '=' case, we also need to append min 
                                                #so that we can avoid error at popping out the min
            self.min.append(value)

    def push(self, values):
        try:
            for value in values:
                self.push_single(value)
            print("Push multiple values")
        except:
            self.push_single(values)
            print("push a single value")
        
    def get_stack_min(self):
        if self.is_empty():
            return None
        else:
            return self.min[-1]
    
    def is_empty(self):
        return len(self.stack)==0
    
    def __str__(self):
        return str(self.stack)
```

```python
mystack = StackWithMin([9, 4, 10, 3, 12, 8])
print(mystack)
print("Min:", mystack.get_stack_min())
print("Pop:", mystack.pop())
print("mystack:", mystack)
print("Min:", mystack.get_stack_min())
value = 2
mystack.push(value)
print("After pushing", value,", mystack:", mystack)
print("Min:", mystack.get_stack_min())
value = [13, 5, 2]
mystack.push(value)
print("After pushing", value,", mystack:", mystack)
print("Min:", mystack.get_stack_min())
print(mystack.min)
print("Pop:", mystack.pop())
print("after pop:", mystack)
print("Min:", mystack.get_stack_min())
popvalue = 1
while 1:
    popvalue = mystack.pop()
    if popvalue == None:
        break
    print("After pop out", popvalue, ", mystack:", mystack)
    print("min:", mystack.min)


```
