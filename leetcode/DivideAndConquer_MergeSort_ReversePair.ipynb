## 493. Reverse Pairs
https://leetcode.com/problems/reverse-pairs/

Given an array nums, we call (i, j) an important reverse pair if i < j and nums[i] > 2*nums[j].

You need to return the number of important reverse pairs in the given array.

Example1:

Input: [1,3,2,3,1]
Output: 2
Example2:

Input: [2,4,3,5,1]
Output: 3
Note:

The length of the given array will not exceed 50,000.
All the numbers in the input array are in the range of 32-bit integer.

```python
class Solution:
    def reversePairs(self, nums: List[int]) -> int:
        
        def merge(left, right, count_rev_pair=False):
            #print('merge:', left, right, count_rev_pair)
            nl, nr = len(left), len(right)
            if count_rev_pair:
                rp_count = 0
                right = [r*2 for r in right] #double operation reserves the sorting property of the array
            else:
                sorted_nums = [] #0 for _ in range(nl + nr)
            il, ir = 0, 0
            while il < nl and ir < nr:
                if left[il] <= right[ir]:
                    if not count_rev_pair:
                        sorted_nums.append(left[il])
                    il += 1
                else: # left[il] > right[ir] 
                    if count_rev_pair:
                        rp_count += nl - il
                    else:
                        sorted_nums.append(right[ir])
                    ir += 1 
            
            if count_rev_pair:
                #print("-", left, right, rp_count)
                return rp_count
            else:
                if ir < nr: #the right still has some remaining elements
                    sorted_nums.extend(right[ir:])
                elif il < nl:
                    sorted_nums.extend(left[il:])
                #print("--", sorted_nums)
                return sorted_nums
        
        
        def sort_and_count(unsorted_nums):
            n = len(unsorted_nums)
            #print('recursive:', unsorted_nums, ', len=', n)
            if n <= 1:
                return unsorted_nums, 0
            else:
                # sort & count rev-pair in the left half and the left right
                sorted_left, count_left = sort_and_count(unsorted_nums[:n//2])
                #print('*l', unsorted_nums, sorted_left, count_left)
                sorted_right, count_right = sort_and_count(unsorted_nums[n//2:])
                #print('*r', unsorted_nums, sorted_right, count_right)
                # sort the whole array
                sorted_nums = merge(sorted_left, sorted_right, False)
                # count the number of important reserve pairs where i < n//2 and j >= n//2
                count_lr = merge(sorted_left, sorted_right, True)
                return sorted_nums, count_left + count_right + count_lr
        
        _, imp_rev_pair_count = sort_and_count(nums)
        return imp_rev_pair_count
            
        
                
        
```
