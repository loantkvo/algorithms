# Two Pointer Technique
Two pointer technique is very effective for searching pairs in a <b>SORTED</b> array 


## 1. Two sum

1.1. Given a sorted array A (sorted in ascending order), having N integers, find if there exists any pair of elements (A[i], A[j]) such that their sum is equal to X. https://www.geeksforgeeks.org/two-pointers-technique/

1.2. Given a sorted array and a number x, find the pair in array whose sum is closest to x
https://www.geeksforgeeks.org/given-sorted-array-number-x-find-pair-array-whose-sum-closest-x/

1.3. Given an integer array A of N elements. You need to find the sum of two elements such that sum is closest to zero. https://practice.geeksforgeeks.org/problems/two-numbers-with-sum-closest-to-zero/0

Input:
The first line of input contains an integer T denoting the number of test cases.  
The first line of each test case is N, denoting the size of array. Each test case consist of a N integers which are the array elements.

Output:
Print the requiired sum which is closest to zero.

Constraints:
1 ≤ T ≤ 100
1 ≤ N ≤ 103
-106 ≤ a[i] ≤ 106

```python
#code
t = int(input())
target = 0 #change this value to x for the questions 1.2
for ti in range(t):
    #read input
    ni = int(input())
    ai = list(map(int, input().strip('\n').split()))
    ai = sorted(ai) #sort the array for two-pointer technique
    
    #initialize
    lp, rp = 0, ni-1 # left pointer and right pointer
    smallest_dist = 2 * 10**6#2**32 if assume that each number is a int32
    while lp<rp:
        dist = ai[lp] + ai[rp] - target
        if dist == 0:
            result = dist + target
            break
        elif dist > 0: #move to the direction that make the sum nearer and nearer to the target
            rp -= 1
        else:
            lp += 1
        
        if abs(dist)<smallest_dist:
            smallest_dist = abs(dist)
            result = dist + target
    print(result)
```

## 2. Three sum
2.1. Count triplets with sum smaller than a given value
https://www.geeksforgeeks.org/count-triplets-with-sum-smaller-that-a-given-value/


2.2. Given an array nums of n integers, are there elements a, b, c in nums such that a + b + c = 0? Find all unique triplets in the array which gives the sum of zero. https://leetcode.com/problems/3sum/

Note:

The solution set must not contain duplicate triplets.

Example:

Given array nums = [-1, 0, 1, 2, -1, -4],

A solution set is:
[
  [-1, 0, 1],
  [-1, -1, 2]
]

```python
class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        n = len(nums)
        if n<3:
            return []
        
        target = 0
        nums = sorted(nums)
        results = []
        for i in range(n):
            if i!=0 and nums[i-1]==nums[i]:
                continue
            
            lp, rp = i+1, n-1
            while lp<rp:
                #print(i, lp, rp)
                tot = nums[i] + nums[lp] + nums[rp]
                if tot==target:
                    results.append([nums[i], nums[lp], nums[rp]]) 
                    while lp<rp and nums[lp]==nums[lp+1]:
                        lp += 1
                    while rp>lp and nums[rp]==nums[rp-1]:
                        rp -= 1
                    lp += 1
                    rp -= 1
                elif tot>target:
                    rp -= 1
                else:
                    lp += 1

        return results
   
```

## 2. Three sum (cont.)
2.3. Given three integer arrays and a “sum”, the task is to check if there are three elements a, b, c such that a + b + c = sum and a, b and c belong to three different arrays. https://www.geeksforgeeks.org/find-three-element-from-different-three-arrays-such-that-that-a-b-c-k/

```python
def threeSum(a1, a2, a3, target):
    n1, n2, n3 = len(a1), len(a2), len(a3)
    results = []
    
    if n1==0 or n2==0 or n3==0:
        return results
    
    a1, a2, a3 = sorted(a1), sorted(a2), sorted(a3)
    
    for i in range(n1):
        if i!=0 and a1[i]==a1[i-1]:
            continue
        j, k = 0, n3-1
        while j<n2 and k>0:
            tot = a1[i] + a2[j] + a3[k]
            if tot > target:
                k -= 1
            elif tot < target:
                j += 1
            else: #tot == target
                results.append([a1[i], a2[j], a3[k]])
                while j<n2-1 and a2[j]==a2[j+1]:
                    j += 1
                j += 1
                while k>0 and a3[k]==a3[k-1]:
                    k -= 1
                k -= 1
    return results
        
    
    
    
```

## 3. Four sum
3.1 https://leetcode.com/problems/4sum/submissions/
Given an array nums of n integers and an integer target, are there elements a, b, c, and d in nums such that a + b + c + d = target? Find all unique quadruplets in the array which gives the sum of target.

Note:

The solution set must not contain duplicate quadruplets.


```python
class Solution:
    def fourSum(self, nums: List[int], target: int) -> List[List[int]]:
        n = len(nums)
        if n<4:
            return []
        
        nums = sorted(nums)
        #print(nums)
        
        results = []
        for i in range(n-3):
            if i!=0 and nums[i-1]==nums[i]:
                continue
            for j in range(i+1, n-2):
                if j!=i+1 and nums[j-1]==nums[j]:
                    continue
                k, h = j+1, n-1
                while k<h:
                    tot = nums[i] + nums[j] + nums[k] + nums[h]
                    #print(tot, i, j, k, h)
                    if tot<target:
                        k += 1
                    elif tot>target:
                        h -= 1
                    else:#tot==target
                        
                        results.append([nums[i], nums[j], nums[k], nums[h]])
                        while k<h and nums[k]==nums[k+1]:
                            k += 1
                        while k<h and nums[h]==nums[h-1]:
                            h -= 1
                        k += 1
                        h -= 1
        return results
```
